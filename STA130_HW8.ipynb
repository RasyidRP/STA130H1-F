## 1. What a Classification Decision Tree Is

A Classification Decision Tree is a machine learning model that uses a tree-like structure that predicts categorical outcomes based on decision rules or questions that each branch has. Each decision rule is based on a specific predictor variable, which is a variable that predicts or explains changes in the outcome variable.

**a) The type of problem a Classification Decision Tree addresses are:**

Classification problems to assign data points to predefined categories. They assign data points to discrete categories, which are distinct, separate groups. They are different from linear regression, which predict continuous outcomes. Examples of real world applications for classification decision trees are: predicting whether or not a patient has a disease based on their medical history, test results, and symptoms; categorizing emails as spam or not based on the content and sender information.

**b) The differences between lies in various different parts:**
1. The outcome variable type of multiple linear regression is continuous, estimating numerical values along a continuous scale. In classification decision trees, the outcome variable type is categorical, meaning each data point is put into predefined discrete categories.
2. The prediction mechanism for multiple linear regression uses a linear equation based on coefficients assigned to each predictor variable. The outcome variable is continuous because it comes from the sums of the weighted contributions of each predictor. For classification decision trees, the prediction mechanism uses a sequential decision-making process that branches out like a tree based on decision rules or questions. Each decision rule stems from predictor variables until it finally reaches a leaf node at the end.
3. Model interpretation for multiple linear regression centers on the coefficients of the linear equation. Each coefficient represents a change in the outcome variable associated with a one-unit change in the corresponding predictor variable, holding other predictors constant. For classification decision trees, the model interpretation focuses on the path and decision rules within the tree structure, leading to a specific prediction.
4. Complexity and Interactions for multiple linear regression assumes a linear relationship between predictors and the outcome. And interactions can be non-linear if they include polynomial terms or transformations of predictor variables. Classification decision trees have a non-linear relationship through a sequential order which allows them to capture interactions between predictors.
5. Overfitting and regularization of multiple linear regression happens when the data has many predictor variables or complex model specifications. To mitigate this, methods like variable selection, dimensionality reduction, and regularization help do the job. For classification decision trees, overfitting happens if the tree has too many branches and can lead to less confident outcomes. A method to mitigate this is by putting a limit to how many data points each branch can have.
6. Model Evaluation for multiple linear regression is usually done using method like R-squared and Mean Squared Error to quantify the model's fit and predictive accuracy. In classification decision trees, we use metrics that are derived from confusion matrices, which summarize the model's performance in classifying data points into correct categories like accuracy, sensitivity, specificity, and precision.

**The difference in the processes between Classification Decision Trees and Multiple Linear Regression are:**

1. Multiple Linear Regression: The prediction mechanism is in linear form. It's an equation that quatifies the relationship between the outcome variable and the predictor variables. It is expressed in this equation:

    Yi = β0 + β1x1i + β2x2i + ... + βnxni + εi

    Where:

    Yi: Represents the outcome variable for a given observation i.

    β0: The intercept term, capturing the average value of Yi when all predictor variables are zero.

    β1, β2, ..., βn: The coefficients associated with each predictor variable (x1i, x2i, ..., xni). Each coefficient quantifies the average change in Yi for a one-unit increase in the corresponding predictor variable, while holding all other predictors constant.

    x1i, x2i, ..., xni: Represent the values of the predictor variables for observation i.

    εi: The error term, accounting for the variability in Yi not explained by the predictor variables.

    The prediction process for multiple linear regression is by plugging in values of the predictor variables for a new observation to the linear equation. The equation then calculates the predicted value of the outcome variable based on the weighted contribution of each predictor variable and the intercept term.

    Example:

    Let's say we want to predict a child's height (Yi) based on their parent's average height (x1i) and their gender (x2i, coded as 0 for female and 1 for male).

    The fitted model is: Yi = 50 + 0.6x1i + 3x2i.

    For a child with a parent's average height of 170 cm and who is male, the predicted height would be: Yi = 50 + 0.6 * 170 + 3 * 1 = 155 cm.
    
2. Classification Decision Trees: Makes predictions by traversing a tree-like structure with parts:
- Nodes: Points where the tree branches based on decision rules.
- Decision Rules: Derived from predictor variables, each decision rule splits the data based on a specific condition.
- Branches: Represent different paths taken based on decision rule outcomes.
- Leaf Nodes: The terminal points of the tree, representing the final prediction category.

    The prediction process for a new observation starts from the root node. The tree starts from the root node and evaluates the first decision rule. Based on the outcome, the tree follows the corresponding branch to the next node, continues based on subsequent decision rules, and eventually ends up at a leaf node. The category assigned to the leaf node becomes the final prediction. 
    
    Example:
- Imagine a decision tree predicting whether a book is hardcover or paperback based on its "Number of Pages" and "Thickness."
- The root node might use the decision rule "Number of Pages > 300."
- If true, the tree branches to a node evaluating "Thickness > 2 cm."
- If also true, the prediction is "Hardcover."
- If false, it might branch to another decision rule or directly predict "Paperback."

## 2. Real-World Applications for Each Metric: Accuracy, Sensitivity, Specificity, Precision

1. **Accuracy** provides a good overall measure of how well a model is performing, but can be misleading when a dataset is imbalanced. It measures the proportion of correctly classified instances. They give broad assessments where the overall performance of a system matters more than its behavior on specific subsets of data. A good example would be a spam email classifier. Let's say 1000 emails are classified as spam, and 950 of them are actually spam emails. Then then model has a 95% accuracy. While high accuracy suggests good overall performance, it does not provide detailed insight into whether it's doing well at avoiding misclassifying critical emails as spam.

2. **Sensitivity** is a good metric when trying to correctly identify all of the positive cases. They are used in scenarios where it's more important to catch all true positives, even when it may result in some false positives. They measure the proportion of actual positives that are correctly identified. A good example would be a wildifre detection system that analyzes satellite imagery to identify potential wildfire breakouts. High sensitivity is crucial because missing one could have devastating consequences. The system should over-detect (which may issue some false alarms) to ensure each potential fire is flagged.

3. **Specificity** is important when you want to avoid false positives. They are used in scenarios where it is important to correctly identify true negatives and avoid false positives, measuring the proportion of true negatives correctly identified. As an example, in a drug testing scenario for athletes, a highly specific test ensures that individuals without the banned substance are correctly identified as negative. This minimizes wrongful accusations of doping. A low specificity would result in many false positives, unfairly implicating innocent athletes and damaging their reputations.

4. **Precision** is important to avoid false positives and when the cost of a false positive is high. It measures the proportion of true positives among the instances predicted as positive. It is used when it is more critical to ensure that positive identifications are truly positive, avoiding "false alarms." A good example is a precision-focused approach is useful for diagnosing a rare disease using a specialized test. Since most people tested are unlikely to have the disease, a positive result must be very likely to be accurate to prevent causing undue stress, unnecessary follow-ups, or ineffective treatments. High precision ensures that when the test claims a person is positive, it is highly confident in that diagnosis.

Overall:

- Accuracy is great for an overall view of a system’s performance but can mask problems if one class is much more common than another (e.g., 95% accuracy may hide poor performance on rare events).
- Sensitivity focuses on catching as many true positives as possible, crucial when missing positives is costly (e.g., missed diseases or emergencies).
- Specificity aims to ensure true negatives are accurately identified, critical when false positives have severe negative repercussions (e.g., wrongful convictions).
- Precision emphasizes the correctness of positive identifications, reducing false positives, which matters in high-stakes decisions where false positives can be problematic (e.g., fraud or rare diseases).

## 9. Differences of interpreting coefficients in linear model regression versus feature importances in decision trees in two to three sentences

In linear regression, coefficients represent the strength and direction of the linear relationship between predictor variables and the target/response variable, quantifying how changes in predictors influence the output. In contrast, feature importances in decision trees reflect how much each feature contributes to reducing impurity at each split, with higher values indicating features that lead to more impactful and effective splits within the tree.
